```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Model names and metric values
models = ['Random Forest', 'SVM', 'MLP']
accuracy = [0.9828, 0.8408, 0.9716]
macro_precision = [0.85, 0.13, 0.79]
macro_recall = [0.71, 0.13, 0.66]
macro_f1 = [0.75, 0.13, 0.66]
weighted_f1 = [0.98, 0.79, 0.97]
```

```python
# Colors for models
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
```

```python
# 1. Accuracy Comparison - Vertical Bar Graph
plt.figure(figsize=(8, 5))
plt.bar(models, accuracy, color=colors)
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2. Macro Metrics Comparison - Line Graph with markers
plt.figure(figsize=(10, 6))
plt.plot(models, macro_precision, marker='o', label='Macro Precision', color=colors[0], linewidth=2)
plt.plot(models, macro_recall, marker='s', label='Macro Recall', color=colors[1], linewidth=2)
plt.plot(models, macro_f1, marker='^', label='Macro F1-Score', color=colors[2], linewidth=2)
plt.title('Macro Metrics Comparison')
plt.ylabel('Score')
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 3. Weighted F1 Comparison - Horizontal Bar Graph
plt.figure(figsize=(8, 5))
plt.barh(models, weighted_f1, color=colors)
plt.title('Weighted F1-Score Comparison')
plt.xlabel('Weighted F1-Score')
plt.xlim(0, 1.1)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
